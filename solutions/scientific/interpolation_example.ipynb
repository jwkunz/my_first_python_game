```python
# Import your libraries here
import numpy as np # Data structures
from scipy.interpolate import interp1d # Algorithms 
import matplotlib.pyplot as plt # Plots
```

```python
# Declare constants and/or the problem definition    

number_of_points_raw = 4
number_of_points_interpolated = 100
```

```python
# Set up working resources

def f(x):
    # f(x) = -5*x^3+8*x^2+0*x+1
    polynomial_coeficients = [-5,8,0,1]
    return np.polyval(polynomial_coeficients,x)
```

```python
# Do the work

# A few points
x_range = np.linspace(-1,1,number_of_points_raw) 
# A lot of points
x_range_interpolated = np.linspace(-1,1,number_of_points_interpolated)
# A function object to perform interpolation
y_interpolator = interp1d(x_range,f(x_range),kind="cubic")
```

```python
# Create a plot to show it off

# Create the plot
plt.figure()
# Plot the lines
plt.plot(x_range_interpolated,y_interpolator(x_range_interpolated),'r-o',label="Interpolated")
plt.plot(x_range,f(x_range),'k-o',label="Original Input")
# Add labels
plt.xlabel("X")
plt.ylabel("Y")
# Add a title
plt.title("Adam's plot")
# Add grid lines
plt.grid()
# Draw it
plt.show()
```
